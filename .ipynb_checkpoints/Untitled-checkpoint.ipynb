```python
import pandas as pd
```

```python
df = pd.read_csv('bahamas_new_listing_details_clean1 - bahamas_new_listing_details_clean1.csv')
```

```python
df
```

```python
df.dropna(subset=["title"] , inplace = True)
```

```python
df.dropna(subset =["img_src"], inplace = True)
```

```python
df
```

```python
df.price
```

```python
df.price
```

```python
df['price']=df['price'].str.replace('$','',regex=False)
```

```python
df.price
```

```python
df['price'] = df['price'].astype(str)
```

```python
df['price_unit'] = df['price'].str.extract(r'(Monthly|Yearly)', expand=False)
df['price'] = df['price'].str.replace(r'[^\d]', '', regex=True).astype(int)
```

```python
df.dropna(subset = ['interior'], inplace =True)
df['interior'] = df['interior'].str.replace(',','')
```

```python
df['interior_sq_ft'] = df['interior'].str.extract(r'(\d+)').astype(int)
```

```python
df['interior_sq_ft']
```

```python
df.dropna(subset = ['exterior'], inplace =True)
```

```python
df['exterior']
```

```python
df['exterior_acres'] = df['exterior'].str.extract(r'(\d+)').astype(int)
```

```python
df['exterior_acres']
```

```python
df["property_types"] = df["link"].str.split("/").str[4]
df["property_types"]
```
